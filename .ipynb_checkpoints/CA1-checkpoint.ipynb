```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("all_data_M_2023.xlsx",sheet_name="All May 2023 data")
```

```python
data.head(20)
```

```python
# As mentioned in the assignment 'You have been tasked with creating visualisations to gain insights into employment in 
# fields involving AI, Data Science and Software Development. '
# So I will filter jobs that I personally consider relevant to the task needed 
relevant_jobs = ['15-1220', ## Computer and Information research scientist highly linked to AI Jobs
                '15-1250', ## combines software and web developers, programmers and testers
                '15-2050' ## data scientist
                ]

## This new dataframe will group the jobs we are interested in all the U.S in all industries
data_cross_us = data[(data['OCC_CODE'].isin(relevant_jobs)) & (data['I_GROUP'] == 'cross-industry')]
data_cross_us.head()
```

```python
data_cross_us.info()
```

```python
##Changed the dataset column tot_emp from a string to numeric, errors will become Nan through 'coerce'
data_cross_us['TOT_EMP'] = pd.to_numeric(data_cross_us['TOT_EMP'], errors='coerce')
data_cross_us[['OCC_TITLE','TOT_EMP']]
```

```python
plt.figure(figsize=(12,8))
plt.pie(data_cross_us['TOT_EMP'],labels=data_cross_us['OCC_TITLE'],autopct='%1.1f%%')
plt.title('Total Employees distribution by Occupation')
plt.show()
```

The previous is a pie plot that I used to understand the overall distribution of the AI, Data Science and Software Development occupations in the United States and across all the industries as part of investigating the dataset. Pie chart in this example works better as it shows percentages of these occupations' distributions. What I've noticed the most was the huge difference in percentages as 90.5% are working in software and web development, programming and testing. While only 8% are working as data scientists and only 1.5% are working as computer and information research scientists.

```python
data_cross_us['A_MEAN'] = data_cross_us['A_MEAN'].astype(str).astype(int)
data_cross_us[['OCC_TITLE','A_MEAN']]
```

```python
plt.figure(figsize=(14,6))
barplot = plt.bar(data_cross_us['OCC_TITLE'],data_cross_us['A_MEAN'],color=['indianred','steelblue','mediumseagreen'])
plt.title("Average Salary by Job title")
plt.xlabel('Job title')
plt.ylabel('Average Salary')
plt.grid()
plt.legend(barplot,data_cross_us['OCC_TITLE'],title="job title")
plt.show()
```

The second plot is a bar plot of the average salary for the 3 broad jobs we have in our data, I did this plot to have an idea if the proportion of jobs would affect the salaries which apparently it doesn't as Computer and Information research scientists make the most, around 157k a year despite being the lowest job category in terms of total employment, software and web devlopers, programmers and testers and Data scientists come next with close salaries, with roughly 130k/year and 119k respectively. Bar plot is a good example of comparing 3 categories (jobs in our example) with a common variable which is the average salary.

```python
# Filtered tha dataset only for relevant columns
filtered_data = data[data['OCC_CODE'].isin(relevant_jobs)]
filtered_data['A_MEAN'] = pd.to_numeric(filtered_data['A_MEAN'], errors='coerce')
filtered_data.info()
```

```python
#Dropped null value
naics_data = filtered_data[['NAICS_TITLE','OCC_TITLE','A_MEAN']]
naics_data=naics_data.dropna()
naics_data.info()
```

```python
##Created a pivot table that shows each industry title, for every occupation and the mean salary
pivot_table = naics_data.pivot_table(index='NAICS_TITLE',columns='OCC_TITLE',values='A_MEAN',aggfunc='mean')
#Then took the top 5 salaries for each job and stored it in a dataframe
cirs_top_5= pivot_table.sort_values(by='Computer and Information Research Scientists', ascending = False).head(5)
ds_top_5= pivot_table.sort_values(by='Data Scientists', ascending = False).head(5)
swdpt_top_5= pivot_table.sort_values(by='Software and Web Developers, Programmers, and Testers', ascending = False).head(5)
pivot_table.head()
```

```python
cirs_top_5
```

```python
swdpt_top_5
```

```python
ds_top_5
```

```python
## the following lines of code are 3 subplots of the 3 previous tables, each job with top 5 paying industries compared to the other jobs
## Subplot was done using matplotlib documentation following the code in the following link:
## https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplots_demo.html
fig, axs = plt.subplots(3, 1,figsize=(20,14))
cirs_top_5.plot(kind='barh',ax=axs[0])
axs[0].set_title('Top 5 Industry Salaries for Information Research Scientists Compared to Software Developers, Programmers, and Testers and Data Scientists')
axs[0].legend(bbox_to_anchor=(1, 1), loc='upper left')


ds_top_5.plot(kind='barh',ax=axs[1])
axs[1].set_title('Top 5 Industry Salaries for Data Scientists Compared to Software Developers, Programmers, and Testers and Information Research Scientists')
axs[1].legend(bbox_to_anchor=(1, 1), loc='upper left')

swdpt_top_5.plot(kind='barh',ax=axs[2])
axs[2].set_title('Top 5 Industry Salaries for Software Developers, Programmers, and Testers Compared to Data Scientists and Information Research Scientists')
axs[2].legend(bbox_to_anchor=(1, 1), loc='upper left')

## same x and y labels for all the plots
for ax in axs.flat:
    ax.set(xlabel='Salary', ylabel='Industry')

for ax in axs.flat:
    ax.label_outer()

plt.tight_layout()
plt.show()
```

the previous 3 subplots are made for comparison reasons, I wanted to have an idea about top 5 paying industries for each position compared to the other 2 positions using a horizontal bar chart, industry on the y axis and salary on the x axis. This type of chart works best when dealing with categorical variable (industry) compared to a salary and showing the 3 occupations at the same time to have a better understanding.
There are many insights we can take from the previous plots, the first one confirms to what we have seen in the previous bar plot where Computer and information research scientists tend to earn much more in general even if it's a high paying industry for the other jobs which highlights the amount of expertise these professionals have and the value they add when it comes to AI employment. 
Another takeaway from the previous visualisation is that some industries tend to invest highly in skilled professionals in AI such as Web search portals,Libraries, and other information services and also media-related services as we find these industries in the top 5 paying industries for our three broad occupations. Also I noticed the variety of the industries that are hugely investing in AI, as there are 9 unique industries paying their employees way above average salaries which confirms the latest directions companies are taking that is AI-Focused in different fields such as Tansit and ground passenger transportation and taxi and limousine services for example.

```python
## Convert relevant columns to numeric
cols_to_convert = ['A_PCT10','A_PCT25','A_MEDIAN','A_PCT75','A_PCT90'] 
filtered_data[cols_to_convert] = filtered_data[cols_to_convert].apply(pd.to_numeric, errors='coerce')
```

```python
### Created a table that groups the 3 occupations, with mean quartile salaries
columns_to_keep = ['OCC_TITLE','A_PCT10','A_PCT25','A_MEDIAN','A_PCT75','A_PCT90'] 
quartile_table = filtered_data[columns_to_keep]
mean_quartile = quartile_table.groupby('OCC_TITLE').agg({
    'A_PCT10':'mean',
    'A_PCT25':'mean',
    'A_MEDIAN':'mean',
    'A_PCT75':'mean',
    'A_PCT90':'mean',
}).reset_index()
mean_quartile
```

```python
boxplot_columns =['A_PCT10', 'A_PCT25', 'A_MEDIAN', 'A_PCT75', 'A_PCT90']
boxplot_data = [
    mean_quartile.loc[0,boxplot_columns].values,
    mean_quartile.loc[1,boxplot_columns].values,
    mean_quartile.loc[2,boxplot_columns].values
]
plt.figure(figsize=(12,8))
plt.boxplot(boxplot_data)

plt.title('Salary distribution for different occupations')
plt.xticks([1,2,3],mean_quartile['OCC_TITLE'])
plt.ylabel('Salary')

plt.tight_layout()
plt.grid()
plt.show()
```

```python
## Challenge Task:
## I have loaded the data jobs for the past 5 years to see trends on AI employment over the past 5 years, that was the further I could go for data reasons
## because earlier years do not have some occupations I'm looking for or else there is not enough documentation ( no occupation profiles section)
data_2019 = pd.read_excel("all_data_M_2019.xlsx",sheet_name="All May 2019 Data")
data_2020 = pd.read_excel("all_data_M_2020.xlsx",sheet_name="All May 2020 data")
data_2021 = pd.read_excel("all_data_M_2021.xlsx",sheet_name="All May 2021 data")
data_2022 = pd.read_excel("all_data_M_2022.xlsx",sheet_name="All May 2022 data")
```

```python
relevant_jobs = ['15-1220', ## Computer and Information research scientist highly linked to AI Jobs
                '15-1250', ## combines software and web developers, programmers and testers
                '15-2050' ## data scientist
                ]
relevant_jobs_19_20 = ['15-1220',
                       '15-1250',
                       '15-2098' ### Data scientists in 2019 and 2020 had different OCC Code
                      ]

## This new dataframe will group the jobs we are interested in all the U.S in all industries
## I did those filters to make sure I get only the 3 occupations on the US Area (nation wide without any other areas and for all industries)
data_cross_us_2019 = data_2019[(data_2019['occ_code'].isin(relevant_jobs_19_20)) & (data_2019['own_code'] == 1235) & (data_2019['area_type'] == 1)]
data_cross_us_2020 = data_2020[(data_2020['OCC_CODE'].isin(relevant_jobs_19_20)) & (data_2020['I_GROUP'] == 'cross-industry') & (data_2020['AREA_TYPE'] == 1)]
data_cross_us_2021 = data_2021[(data_2021['OCC_CODE'].isin(relevant_jobs)) & (data_2021['I_GROUP'] == 'cross-industry')& (data_2020['AREA_TYPE'] == 1)]
data_cross_us_2022 = data_2022[(data_2022['OCC_CODE'].isin(relevant_jobs)) & (data_2022['I_GROUP'] == 'cross-industry')& (data_2020['AREA_TYPE'] == 1)]
## this dataframe I already predefined in the the fourth cell
data_cross_us_2023 = data_cross_us

## I took the relevant columns here
data_cross_us_2019 = data_cross_us_2019[['occ_code','occ_title','a_mean','tot_emp']]
data_cross_us_2020 = data_cross_us_2020[['OCC_CODE','OCC_TITLE','A_MEAN','TOT_EMP']]
data_cross_us_2021 = data_cross_us_2021[['OCC_CODE','OCC_TITLE','A_MEAN','TOT_EMP']]
data_cross_us_2022 = data_cross_us_2022[['OCC_CODE','OCC_TITLE','A_MEAN','TOT_EMP']]
data_cross_us_2023 = data_cross_us_2023[['OCC_CODE','OCC_TITLE','A_MEAN','TOT_EMP']]
```

```python
data_cross_us_2019
## I changed some column values because data scientists in 2019 had a different name
data_cross_us_2019['occ_title']=data_cross_us_2019['occ_title'].str.replace('Data Scientists and Mathematical Science Occupations, All Other','Data Scientists')
data_cross_us_2019['occ_code']=data_cross_us_2019['occ_code'].str.replace('15-2098','15-2050')
## added a year column
data_cross_us_2019['Year']=2019
## I changed column names to keep data integrity for further manipulation
data_cross_us_2019.rename(columns={'occ_code':'OCC_CODE',
                                   'occ_title' : 'OCC_TITLE',
                                   'a_mean' : 'A_MEAN',
                                   'tot_emp' : 'TOT_EMP'
                                  },inplace=True)
data_cross_us_2019
```

```python
## Same as the previous dataframe
data_cross_us_2020
data_cross_us_2020['OCC_TITLE']=data_cross_us_2020['OCC_TITLE'].str.replace('Data Scientists and Mathematical Science Occupations, All Other','Data Scientists')
data_cross_us_2020['OCC_CODE']=data_cross_us_2020['OCC_CODE'].str.replace('15-2098','15-2050')
data_cross_us_2020['Year']=2020
data_cross_us_2020
```

```python
data_cross_us_2021['Year']=2021
data_cross_us_2021
```

```python
data_cross_us_2022['Year']=2022
data_cross_us_2022
```

```python
data_cross_us_2023['Year']=2023
data_cross_us_2023
```

```python
## grouped the 5 dataframes into one single dataframe
data_2019_2023 = pd.concat([data_cross_us_2019,data_cross_us_2020,data_cross_us_2021,data_cross_us_2022,data_cross_us_2023],ignore_index=True)
data_2019_2023
```

```python
## subset for each occupation for easier line plots
ds_2019_2023 = data_2019_2023[data_2019_2023['OCC_TITLE']=='Data Scientists']
swdpt_2019_2023 = data_2019_2023[data_2019_2023['OCC_TITLE']=='Software and Web Developers, Programmers, and Testers']
cirs_2019_2023 = data_2019_2023[data_2019_2023['OCC_TITLE']=='Computer and Information Research Scientists']
```

```python
fig, axs = plt.subplots(2, 1,figsize=(12,12))

## each occupation will have its own line plot and label
axs[0].plot(ds_2019_2023['Year'],ds_2019_2023['A_MEAN'],label='Data Scientists')
axs[0].plot(swdpt_2019_2023['Year'],swdpt_2019_2023['A_MEAN'],label='Software and Web Developers, Programmers, and Testers')
axs[0].plot(cirs_2019_2023['Year'],cirs_2019_2023['A_MEAN'],label='Computer and Information Research Scientists')
axs[0].set_xlabel('Year')
axs[0].set_ylabel('Average Mean Salary')
axs[0].set_title('Salary Trends by Occupation Over Time')
axs[0].legend(bbox_to_anchor=(1,1),loc='upper left')

axs[1].plot(ds_2019_2023['Year'],ds_2019_2023['TOT_EMP'],label='Data Scientists')
axs[1].plot(swdpt_2019_2023['Year'],swdpt_2019_2023['TOT_EMP'],label='Software and Web Developers, Programmers, and Testers')
axs[1].plot(cirs_2019_2023['Year'],cirs_2019_2023['TOT_EMP'],label='Computer and Information Research Scientists')
axs[1].set_xlabel('Year')
axs[1].set_ylabel('Total employment')
axs[1].set_title('Total employment by Occupation Over Time')
axs[1].legend(bbox_to_anchor=(1,1),loc='upper left')

plt.show()
```

As we can see from the previous two plots, we notice that AI employment is growing in terms of employment especially for software and web developers, programmers and testers(SWDPT) and data scientists. The total number of employment went from 1.7 million and 30k in 2019 to 2.1 millions and 192k in 2023 for the two occupations respectively. This increase in employment workforce was actually accompanied with major pay increases for all AI employment occupations. For example, computer and information research scientists went from 127k/year in 2019 to to 157k/year in 2023, same for data scientists who used used to make 100k/year in 2019 to 119k/year in 2023, the same for SWDPT that used to went from 106k/year inn 2019 to 130k/year in 2023.
As a conclustion we notice that the AI employment is moving forward throughout the years and companies are more and more investing in this field by getting more skilled people each year and motivating them through consistently increasing salaries.
