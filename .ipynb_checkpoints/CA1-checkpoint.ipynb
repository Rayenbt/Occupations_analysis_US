```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("all_data_M_2023.xlsx",sheet_name="All May 2023 data")
```

```python
data.head(20)
```

```python
# As mentioned in the assignment 'You have been tasked with creating visualisations to gain insights into employment in 
# fields involving AI, Data Science and Software Development. '
# So I will filter jobs that I personally consider relevant to the task needed 
relevant_jobs = ['15-1220', ## Computer and Information research scientist highly linked to AI Jobs
                '15-1250', ## combines software and web developers, programmers and testers
                '15-2050' ## data scientist
                ]

## This new dataframe will group the jobs we are interested in all the U.S
filtered_data = data[data['OCC_CODE'].isin(relevant_jobs)]
filtered_data.head(15)
```

```python
filtered_data.columns
```

```python
##Changed the dataset column tot_emp from a string to numeric, errors will become Nan through 'coerce'
filtered_data.loc[:,'TOT_EMP'] = pd.to_numeric(filtered_data['TOT_EMP'], errors='coerce')
#group filtered data based on the job titles and we do the sum
total_employees = filtered_data.groupby(['OCC_TITLE'])['TOT_EMP'].sum()
total_employees
```

```python
plt.figure(figsize=(12,8))
plt.pie(total_employees,labels=total_employees.index,autopct='%1.1f%%')
plt.title('Total Employees distribution by Occupation')
plt.show()
```

The previous plot is a pie plot that I've used to see the overall distribution of the jobs in the dataset in percentages as part of investigating the dataset, pie plot works best in this case to have an idea about the weight of each job we have.
I noticed that a huge amount of people are into software and web development with 90.2%  while only 8.2% are into data science however I found only 1.2% are into computer and information reasearch science.
